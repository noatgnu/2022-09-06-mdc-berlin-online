```python
# function for writing out data associated to multiple years
def yearly_data_csv_writer(start_year, end_year, data):
    """
    Write out yearly data associated to a range of years
    
    start_year -- the starting year of the range
    end_year -- the ending year of the range
    data -- the pandas dataframe of interests
    """
    for year in range(start_year, end_year+1):
        tempt_data = data[data["year"]== year]
        if tempt_data.empty:
            print("There is no data for year" + str(year))
        else:
            tempt_data.to_csv("data/yearly_data_"+str(year)+".csv")
    
```

```python
# a function to subset a dataframe for a given range of years
def subset_year_range(data, start_year=1977, end_year=2002):
    """
    Create a subset of the data of interests containing only specific range of years
    start_year -- starting year
    end_year -- ending year
    data -- a pandas dataframe of interests
    """
    return data[data["year"].isin(list(range(start_year, end_year+1)))]
```

```python
def checking_number_of_column(data):
    """
    Checking number of column from a dataframe
    data -- dataframe of interests
    """
    if len(data.columns) == 9:
        print("Dataframe has 9 columns")
    elif len(data.columns) < 9:
        print("Dataframe has less than 9 columns")
    else:
        print("Dataframe has more than 9 columns")
    
```

```python
# create a function that would take a range of year as parameters and save plots into file
def yearly_plot(data, filepath, start_year=1977, end_year=2002):
    """
    Create a plot for each year from data frame where x = weight and y = hindfoot_length
    data -- dataframe of interests
    filepath -- path to the folder to store our plots
    start_year -- starting year
    end_year -- ending year
    """
    
    for year in range(start_year, end_year+1):
        temp_df = data[data["year"]== year]
        if temp_df.empty:
            print("No data found in year", year)
        else:
            plot = (p9.ggplot(temp_df, p9.aes(x="weight", y="hindfoot_length"))
                   + p9.geom_point(alpha=0.1) + p9.theme_bw()
                   )
            plot.save(filepath + str(year)+".png")
```
